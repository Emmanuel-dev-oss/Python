```python
#String Manipulation

"""Convert A String To Lowercase"""
text = "HELLO WORLD"
lowercase_text = text.lower()
print(f"Original text: {text}")
print(f"Lowercase text: {lowercase_text}")
print("                      ")


"""Convert A String To Uppercase"""
text = "hello world"
uppercase_text = text.upper()
print(f"Original text: {text}")
print(f"Uppercase text: {uppercase_text}")
print("                      ")


"""String Replacement"""
text = "I'm An Approved Developer"
new_text = text.replace("Approved", "Unique")
print(f"Original text: {text}")
print(f"New text: {new_text}")
print("                      ")


"""String Counting"""
text = "I'm An Approved Developer"
count = text.count("o")
print(f"Original text: {text}")
print(f"Count of o: {count}")
print("                      ")


"""String Spliting"""
text = "I'm An Approved Developer"
words = text.split()
print(f"Original text: {text}")
print(f"Words: {words}")
print("                      ")

"""String Stemming and Lemmatization"""
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer
nltk.download('wordnet')
text = "Running with little children makes sadness disappear"
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
stemmed_words = [stemmer.stem(word) for word in text.split()]
lemmatized_words = [lemmatizer.lemmatize(word) for word in text.split()]
print(f"Original text: {text}")
print(f"Stemmed words: {stemmed_words}")
print(f"Lemmatized words: {lemmatized_words}")
```


#Secure Banking System Pseudocode

##Variables
- `account_number`: unique account identifier
- `pin`: 4-digit personal identification number
- `balance`: current account balance
- `transaction_history`: record of all transactions

Main Program
*Initialize Account*
1. Set `account_number` to a unique value
2. Set `pin` to a secure value
3. Set `balance` to 0
4. Initialize `transaction_history` as an empty list

Display Menu
1. Print "Welcome to Secure Banking!"
2. Print "Please select an option:"
3. Print "1. Deposit"
4. Print "2. Withdraw"
5. Print "3. Check Balance"
6. Print "4. Transaction History"
7. Print "5. Exit"

Get User Input
1. Ask user to take their choice of action (1, 2, 3, 4, or 5)

Process User Input

*Deposit*

1. Ask user to enter amount to deposit
2. Verify user's identity using `pin`
3. Add amount to `balance`
4. Append transaction to `transaction_history`
5. Print "Deposit successful!"


*Withdraw*

1. Ask user to enter amount to withdraw
2. Verify user's identity using `pin`
3. Check if `balance` is sufficient
4. If sufficient, subtract amount from `balance`
5. Append transaction to `transaction_history`
6. Print "Withdrawal successful!"

*Check Balance*

1. Verify user's identity using `pin`
2. Print "Your current balance is: " + `balance`

*Transaction History*

1. Verify user's identity using `pin`
2. Print `transaction_history`


*Exit*

1. Print "Thank you for using Secure Banking!"
2. Exit program

Error Handling
*Invalid Input*

1. If user enters an invalid choice (not 1, 2, 3, 4, or 5)
2. Print "Invalid choice. Please try again."

*Invalid PIN*

1. If user enters incorrect `pin`
2. Print "Invalid PIN. Please try again."

*Insufficient Balance*

1. If user tries to withdraw more than their

current balance
2. Print "Insufficient balance. Please try again."
